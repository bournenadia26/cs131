```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import CondensedNearestNeighbour, TomekLinks, OneSidedSelection, EditedNearestNeighbours, RepeatedEditedNearestNeighbours, AllKNN
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/bournenadia26/cs131/refs/heads/main/finalv.csv")
X = df[['radius2', 'radius3']]
y = df['diagnosis']

#q11 (oversampling)
X_ROS, y_ROS = RandomOverSampler(random_state=1).fit_resample(X, y) #Random Over-sampling
X_SMOTE, y_SMOTE = SMOTE(random_state=1).fit_resample(X, y) #SMOTE

#q12 (undersampling)
X_CNN, y_CNN = CondensedNearestNeighbour().fit_resample(X, y) #Condensed Nearest Neighbours
X_TL, y_TL = TomekLinks().fit_resample(X, y) #Tomek Links
X_OSS, y_OSS = OneSidedSelection().fit_resample(X, y) #One-sided Selection
X_ENN, y_ENN = EditedNearestNeighbours().fit_resample(X, y) #Edited Nearest Neighbours
X_RENN, y_RENN = RepeatedEditedNearestNeighbours().fit_resample(X, y) #Repeated ENN
X_allKNN, y_allKNN = AllKNN().fit_resample(X, y) #All k-NN

# source: https://www.geeksforgeeks.org/matplotlib-pyplot-subplots-in-python/
#q13 (visualize)
def plot(X_new, y_new, title):
    df_plot = X_new.copy()
    df_plot['diagnosis'] = y_new

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.scatterplot(data=df, x='radius2', y='radius3', hue='diagnosis')
    plt.title("Original")
    plt.xlabel("radius2")
    plt.ylabel("radius3")
    plt.legend(title='Diagnosis')

    plt.subplot(1, 2, 2)
    sns.scatterplot(data=df_plot, x='radius2', y='radius3', hue='diagnosis')
    plt.title(title)
    plt.xlabel("radius2")
    plt.ylabel("radius3")
    plt.legend(title='Diagnosis')

    plt.tight_layout()
    plt.show()

#oversampling
plot(X_ROS, y_ROS, "After Random OverSampling")
plot(X_SMOTE, y_SMOTE, "After SMOTE")

#undersampling
plot(X_CNN, y_CNN, "After Condensed Nearest Neighbours")
plot(X_TL, y_TL, "After Tomek Links")
plot(X_OSS, y_OSS, "After One-sided Selection")
plot(X_ENN, y_ENN, "After ENN")
plot(X_RENN, y_RENN, "After Repeated ENN")
plot(X_allKNN, y_allKNN, "After All k-NN")
```
